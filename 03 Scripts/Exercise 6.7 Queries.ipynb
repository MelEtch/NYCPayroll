```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
#define path to prepared data folder
preparedDataPath= r'C:\Users\user\nycpayrollanalysis\02 Data\prepareddata'
```

```python
#import data
goodcols=["Agency Name","Regular Gross Paid"]
df=pd.read_csv(os.path.join(preparedDataPath,'payroll_clean.csv'),usecols=goodcols)
```

```python
df.head()
```

```python
df['RGP_mean']=df.groupby('Agency Name')['Regular Gross Paid'].transform('mean')
```

```python
df.head()
```

```python
df_small=df.drop(columns='Regular Gross Paid')
```

```python
df_small_nodup=df_small.drop_duplicates()
```

```python
df_small_nodup.head()
```

```python
df_small_nodup.shape
```

```python
df_sort=df_small_nodup.sort_values(by=['RGP_mean'])
```

```python
agency=df_sort['Agency Name']
avg_pay=df_sort['RGP_mean']
```

```python
plt.bar(agency, avg_pay)

plt.title('Average Gross Paid by Agency')
plt.xlabel('Agency Name',fontsize=12)
plt.ylabel('Average Pay', fontsize=12)
plt.show()
```

```python
#export to view in excel
df_sort.to_csv(os.path.join(preparedDataPath, 'agencyavgpay.csv'))
```
